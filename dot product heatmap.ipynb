```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating 10 random 3D vectors
num_vectors = 10
dimension = 3
vectors = np.random.randn(num_vectors, dimension)

# Normalizing vectors (unit length)
norms = np.linalg.norm(vectors, axis=1, keepdims=True)
unit_vectors = vectors / norms

# Computing dot product matrix
dot_matrix = np.dot(unit_vectors, unit_vectors.T)

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(dot_matrix, annot=True, cmap="inferno", vmin=-1, vmax=1, fmt=".2f",
            square=True, cbar_kws={"label": "cos(θ)"})
plt.title("Dot Product Heatmap of Normalized Random Vectors")
plt.xlabel("Vector Index")
plt.ylabel("Vector Index")
plt.show()
```

###  Interpretation of the Dot Product Heatmap

This heatmap shows the dot products between 10 randomly generated, **normalized 3D vectors**. The purpose of this visualization is to understand how vector direction relationships (alignment, orthogonality, opposition) show up in matrix form — and how we can intuitively interpret them using color.

---

####  What does each value represent?

Since all vectors$$ \vec{u}_i $$ are normalized to unit length, the dot product between any two vectors $$\vec{u}_i$$ and $$\vec{u}_j$$ is given by:

$$
\vec{u}_i \cdot \vec{u}_j = \|\vec{u}_i\| \|\vec{u}_j\| \cos(\theta_{ij}) = \cos(\theta_{ij})
$$

This means each matrix entry directly represents the **cosine of the angle** between the two vectors. The result always lies in the range:

- \( +1 \) → perfectly aligned (angle = 0°)  
- \( 0 \) → perpendicular (angle = 90°)  
- \( -1 \) → opposite directions (angle = 180°)  

---

#### 🔹 Diagonal is always 1.00

This is because any vector is always perfectly aligned with itself:

$$
\vec{u}_i \cdot \vec{u}_i = \|\vec{u}_i\|^2 = 1
$$

This results in all diagonal values being exactly **1.00**, which shows up as the **brightest yellow** squares along the diagonal in the heatmap.

---

#### 🔹 The matrix is symmetric

Notice how the matrix is a mirror image across the diagonal — this is not accidental.

Dot product is commutative:

$$
\vec{u}_i \cdot \vec{u}_j = \vec{u}_j \cdot \vec{u}_i
$$

So the matrix is always **symmetric**:

$$
D_{ij} = D_{ji}
$$

This symmetry is clearly visible in the heatmap. Every value above the diagonal has an identical counterpart below it.

---

#### 🔹 Interpretation via colors (Inferno colormap)

The heatmap uses the **Inferno** colormap, where:

- **Bright yellow** → high dot product (strong alignment)  
- **White/orange tones** → dot product near zero (perpendicular)  
- **Dark purple to black** → negative dot product (opposite directions)  

For example:
- Entry (4,5) and (5,4) ≈ **-0.91**: strongly opposite directions (darkest cells)  
- Entry (0,9) ≈ **0.61**: moderately aligned  
- Entry (7,6) ≈ **-0.91**: strongly opposing  

This visual cue makes it easy to spot directional relationships **at a glance**.

---

####  Matrix Summary (Mathematically)

Let $$\mathbf{U} \in \mathbb{R}^{10 \times 3}$$ be the matrix of normalized vectors. Then:

- Dot product matrix:  
  $$
  \mathbf{D} = \mathbf{U} \mathbf{U}^\top
  $$

- Properties:  
  $$
  D_{ij} = \cos(\theta_{ij}), \quad D_{ii} = 1, \quad D_{ij} = D_{ji}
  $$

This matrix acts like a "map" of how similar each vector is to every other — just in terms of **direction**, not magnitude.
